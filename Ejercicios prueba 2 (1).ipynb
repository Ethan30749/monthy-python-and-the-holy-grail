```python
from random import randint

Matriz=[
    [],
    [],
    [],
]

M=len(Matriz)
N=3
for i in range (M):
    for j in range(3):
        Matriz[i].append(randint(-50, 100))

        
        

```

```python
M=len(Matriz)
N=len(Matriz[0])
posi_counter=0
nega_counter=0
neut_counter=0
for i in range(M):
    for j in range (N):
        value=Matriz[i][j]
        if value < 0:
            nega_counter+=1
        elif value > 0:
            posi_counter+=1
        else:
            neut_counter+=1
print("Matriz estudiada: ", Matriz)
print("valores positivos: ", posi_counter)
print("valores negativos: ", nega_counter)
print("valores neutros: ", neut_counter)
```

```python
if posi_counter > nega_counter:
    print("hay mas valores positivos que negativos")
else:
    print("hay mas valores negativos que positivos")
```

```python
nega_sum=0
posi_sum=0
for i in range(M):
    for j in range (N):
        value=Matriz[i][j]
        if value <0:
            nega_sum+=value
        elif value > 0:
            posi_sum+=value
print("sumatoria valores positivos: ", posi_sum)
print("sumatoria valores negativos: ", nega_sum)

if posi_sum > (nega_sum*-1):
    print("la sumatoria positiva es mayor que la negativa")
```

```python
def perfecto(A):
    sum=0
    if A > 0:
        for i in range (1, A):
            if A % i == 0:
                sum+=i
    if sum == A:
        return True

```

```python
suma_perfecto=0

for i in range(M):
    for j in range (N):
        value=Matriz[i][j]
        if perfecto(value) == True:
            suma_perfecto+=value
print("sumatoria de todos los valores perfectos en la matriz:",suma_perfecto)
```

```python
from random import randint
Gato= [
    [],
    [],
    [],
]
M=3
N=3
for i in range (M):
    for j in range (N):
        azar=randint(1,2)
        if azar == 1:
            Gato[i].append("X")
        else:
            Gato[i].append("O")
print(Gato)
O_count=0
X_count=0
for i in range (M):
    for j in range (N):
        gatos=Gato[i][j]
        if gatos == "O":
            O_count+=1
        else:
            X_count+=1
print("Hay un total de",O_count, "Os en el gato \n existen un total de",X_count,"Xs en el gato")
```

```python
with open("texto.txt","r") as archivo:
    lectura=archivo.read()
print(lectura)
vocales=[["a","A","á","Á"],["e","E","é","É"],["i","I","í","Í"],["o","O","ó","Ó"],["u","U","ú","Ú"]]
def diccionario(M):
    
    dict={}
    a_count=0
    e_count=0
    i_count=0
    o_count=0
    u_count=0
    vocal_count=0
    consonant_count=0
    
    
    for i in M:
        if i in vocales[0]:
            a_count+=1

        elif i in vocales[1]:
            e_count+=1
            
        elif i in vocales[2]:
            i_count+=1
        elif i in vocales[3]:
            o_count+=1
        elif i in vocales[4]:
            u_count+=1
        else:
            consonant_count+=1
            
            
    dict["cantidades de a"]=a_count
    dict["cantidades de e"]=e_count
    dict["cantidades de i"]=i_count
    dict["cantidades de o"]=o_count
    dict["cantidades de u"]=u_count
    dict["cantidades de consonantes"]=consonant_count
    
    
    return dict

diccionario(lectura)
```

```python
def palindromo(M):
    palabra=""
    for i in M:
        if i != " ":
            palabra+=i
    if palabra == str(palabra[::-1]):
        return True
    
M=input("ingrese una palabra: ")
print(palindromo(M))
```
