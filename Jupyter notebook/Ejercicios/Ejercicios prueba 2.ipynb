```python
from random import randint

Matriz=[
    [],
    [],
    [],
]

M=len(Matriz)
N=3
for i in range (M):
    for j in range(3):
        Matriz[i].append(randint(-50, 100))

        
        

```

```python
M=len(Matriz)
N=len(Matriz[0])
posi_counter=0
nega_counter=0
neut_counter=0
for i in range(M):
    for j in range (N):
        value=Matriz[i][j]
        if value < 0:
            nega_counter+=1
        elif value > 0:
            posi_counter+=1
        else:
            neut_counter+=1
print("Matriz estudiada: ", Matriz)
print("valores positivos: ", posi_counter)
print("valores negativos: ", nega_counter)
print("valores neutros: ", neut_counter)
```

```python
if posi_counter > nega_counter:
    print("hay mas valores positivos que negativos")
else:
    print("hay mas valores negativos que positivos")
```

```python
nega_sum=0
posi_sum=0
for i in range(M):
    for j in range (N):
        value=Matriz[i][j]
        if value <0:
            nega_sum+=value
        elif value > 0:
            posi_sum+=value
print("sumatoria valores positivos: ", posi_sum)
print("sumatoria valores negativos: ", nega_sum)

if posi_sum > (nega_sum*-1):
    print("la sumatoria positiva es mayor que la negativa")
```

```python
matriz=[
    [2,3,4],
    [5,3,1],
    [4,8,4],
]
valores_unicos=[]
valores_repetidos=[]
for i in range(len(matriz)):
    for j in range (len(matriz[0])):
        valor=matriz[i][j]
        if valor not in valores_unicos:
            valores_unicos.append(valor)
        else:
            if valor not in valores_repetidos:
                valores_repetidos.append(valor)
print(matriz)
print(valores_unicos)
print(valores_repetidos)
print("hay un total de",len(valores_unicos)-len(valores_repetidos), "valores unicos en la matriz")
```

```python
def perfecto(A):
    sum=0
    if A > 0:
        for i in range (1, A):
            if A % i == 0:
                sum+=i
    if sum == A:
        return True

```

```python
suma_perfecto=0

for i in range(M):
    for j in range (N):
        value=Matriz[i][j]
        if perfecto(value) == True:
            suma_perfecto+=value
print(suma_perfecto)
```
